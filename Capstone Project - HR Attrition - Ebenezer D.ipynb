### Age

```python
# fill null values in age with avg of Mean and Median = round off to 37

db_copy['Age'].fillna(37, inplace=True)

#convert Age to int 

db_copy['Age'] = db_copy['Age'].astype(int)

#Letting outlers be as we can see in cell 22 not much difference b/n Mean and Median 
```

### Distance from home to work

```python
# fill null values in H2W with Mean+Median/2 (let us round off to 10)

db_copy['HomeToWork'].fillna(10, inplace=True)

#convert H2W to int 

db_copy['HomeToWork'] = db_copy['HomeToWork'].astype(int)

#Letting outliers be as we can see in cell 22 not much difference b/n Mean and Median and I am considering 10 which is 
# closer to Median
```

### Hours in a week

```python
# fill null values in Hours in a week with Median

db_copy['HourlnWeek'].fillna(db_copy['HourlnWeek'].median(), inplace=True)

#convert Hours in a week to int 

db_copy['HourlnWeek'] = db_copy['HourlnWeek'].astype(int)

plt.figure(figsize=(8, 6))
sns.boxplot(x= db_copy['HourlnWeek'])
plt.title('HourlnWeek')
plt.xlabel('HourlnWeek')
plt.show()
```

```python
Q1 = db_copy['HourlnWeek'].quantile(0.25)
Q3 = db_copy['HourlnWeek'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)

print('-'*120)

below_outliers = db_copy[db_copy['HourlnWeek'] < lower_bound]
above_outliers = db_copy[db_copy['HourlnWeek'] > upper_bound]

count_below_outliers = len(below_outliers)
count_above_outliers = len(above_outliers)

print("Count of values below the lower outlier threshold:", count_below_outliers)
print("Count of values above the upper outlier threshold:", count_above_outliers)

# 5 employees are working less hours and 5 long hours, 
# I am considering them since they are not affecting the avg mean or median.
```

```python
below_outliers = db_copy[db_copy['HourlnWeek'] < lower_bound]['HourlnWeek']
above_outliers = db_copy[db_copy['HourlnWeek'] > upper_bound]['HourlnWeek']

outlier_values = list(below_outliers) + list(above_outliers)

print("List of outlier values:")
print(outlier_values)
```
